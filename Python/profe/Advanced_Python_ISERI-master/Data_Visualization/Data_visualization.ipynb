## Generando gráficos

En la siguiente celda se muestra un programa básico en el que se tiene una lista con valores de $x^2$  y se desean graficar los puntos. 

| x | y | 
| --- | --- | 
| 1 | 1 |
| 2 | 4 |
| 3 | 9 | 
| 4 | 16 | 
| 5 | 25 |

```python
import matplotlib.pyplot as plt

%matplotlib inline

x = [1,2,3,4,5]
y = [1,4,9,16,25]

fig,ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

Lo primero que se tiene que hacer es importar el módulo `pyplot` utilizando el alias `plt` para no tener que repetir `pyplot`, simplemente se pone `plt` y listo.  Se han creado dos listas que almacenan los datos de la tabla. La parte sustancial es cuando se llama al método `subplots` que  ayuda apoder crear uno o varias gráficas en la misma figura. La variable `fig` es el cojunto de gráficos que se han generado. La variable `ax` representa en específico la gráfica donde están los datos de `x` y `y`. Cuando se utiliza el método `plot()` se genera el gráfico con la informaciń que se mande como argumentos, en este caso sería la lista `x` y la lista `y` y una vez que se tienen y la gráfica se debe de mostrar con `show()` de lo contrario la gráfica simplemente no aparecerá .

### Agregando título al grafíco y texto para los ejes coordenados

Aunque se ha mostrado el gráfico se pueden modificar  varias cosas que podrían ayudar al lector a poder interpretar la gráfica como el título de la imagen, además del texto que se puede agregar para cada eje. para esto existen los metodos `.set_title()`, `.set_xlabel()` y `.set_ylabel()`, los tres métodos aceptan los mismos parámetros, el primero es el texto que se quiere agregar y segundo es el tamaño del texto, por ejemplo:

`ax.set_title("x^2",fontsize = 24)`

`ax.set_xlabel("valor", fontsize = 24)`

`ax.set_ylabel("x", fontsize = 36)`

```python
import matplotlib.pyplot as plt

%matplotlib inline

x = [1,2,3,4,5]
y = [1,4,9,16,25]

fig,ax = plt.subplots()
ax.plot(x,y)
ax.set_title("x^2",fontsize = 24)
ax.set_xlabel("x", fontsize = 24)
ax.set_ylabel("y", fontsize = 36)

plt.show()
```

Hasta ahorita se ha podido graficar el conjunto de puntos indicados a través de dos listas, sin embargo, dentro de la gráfica no se observa el punto de manera explícita, supong ahora que quiere hacer énfasis en algún punto en específico sólo para ejemplificar se quiere enfatizar en el punto `(4,16)`, se debe de hacer uso del método `.scatter(valor_x, valor_y,tamaño)` donde `valor_x` es la coordenada del punto en el eje `x` y `valor_y` es la coordenada del putno en el eje `y` y el tamaño en que se quiere que se haga el punto está asociado al parámetro `tamaño`. En la siguiente celda se ejemplifica el uso de este método.


```python
import matplotlib.pyplot as plt

%matplotlib inline

x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.style.use('seaborn')
fig,ax = plt.subplots()
ax.plot(x,y,linewidth = 4)
ax.set_title("x^2",fontsize = 24)
ax.set_xlabel("x", fontsize = 24)
ax.set_ylabel("y", fontsize = 36)
ax.tick_params(axis = 'both', labelsize = 14)
ax.scatter(4,16, s = 200)

plt.show()
```

En caso que se quisieran mostrar todos los puntos y además cambiar el color en el método `scatter()` se debe de agregar toda la lista de los puntos tanto en el eje `x` como en el eje `y`, y se agrega el atributo `c = 'red'` que corresponde al color rojo. El uso de estos nuevos elementos se visualiza en la siguiente celda

```python
import matplotlib.pyplot as plt

%matplotlib inline

x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.style.use('seaborn')
fig,ax = plt.subplots()
ax.plot(x,y,linewidth = 4)
ax.set_title("x^2",fontsize = 24)
ax.set_xlabel("x", fontsize = 24)
ax.set_ylabel("y", fontsize = 36)
ax.tick_params(axis = 'both', labelsize = 14)
ax.scatter(x,y,c='red', s = 200)

plt.show()
```

### Generación de datos y curvas suaves

Suponga ahora que se quiere incrementar el número de puntos para tener como resultado una curva mas suave, para esto se tiene que utilizar un ciclo `for` sobre una lista con `n` puntos, generados por `range` en la siguiente celda se puede observar el desarrollo del programa, sólo se necesitan el punto inicial y el punto final.

El desarrollo del programa se observa en la siguiente celda.

```python
import matplotlib.pyplot as plt

%matplotlib inline

x_init = 0;
x_fin = 500;


x_values = range(x_init,x_fin)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn')
fig,ax = plt.subplots()
ax.plot(x_values,y_values,linewidth = 4)
ax.set_title("x^2",fontsize = 24)
ax.set_xlabel("x", fontsize = 24)
ax.set_ylabel("y", fontsize = 36)
ax.tick_params(axis = 'both', labelsize = 14)


plt.show()
```

## Gráficos de barras, histogramas

Para esta sección realizaremos una clase que se llame `dado` que simulará el efecto de tirar un dado de seis caras. La clase está en la siguiente celda

```python
from random import randint

class Dado:
    """Clase que simula un dado"""
    
    def __init__(self,num_lados = 6):
        '''De manera predeterminada el dado tiene 6 caras'''
        self.num_caras = num_lados
        
    def tira_dados(self):
        '''Regresa el valor después de lanzar el dado'''
        return randint(1,self.num_caras)
```

La función `randint(init_val, final_val)` regresa un valor aleatorio entre el mínimo `init_val` y el valor final `final_val`, se sugiere que si tiene dudas en la implementación de la clase revise el Notebook Jupyter que hace referencia a estos aspectos, se ecuentra en el mismo repositorio que este mismo documento.

Se quiere simular el resultado de tirar el dado 1000 veces y poder representar el resultado y hacer el análisis correspondiente, para la simulación de las 100 veces se muestra el siguiente código

```python
miDado = Dado()
resultados = []
for i in range(0,1000,1):
    resultados.append(miDado.tira_dados())
print(resultados)
```

Una vez que se han obtenido los datos, es necesario comenzar a contar la frecuencia con la que aparecen los datos y graficarlos, esto se hace através de un histograma, pero antes de hacer el gráfico, es necesario hacer el procesamiento de los datos

```python
frecuencias = []
for valor in range(1,miDado.num_caras+1):
    frecuencia = resultados.count(valor)
    frecuencias.append(frecuencia)
print(frecuencias)
```

En esta etapa se está contanto cuántas veces se obtuvo como resultado 1, cuántas veces se obtuvo como resultado 2 y así sucesivamente y se guardan los valore en la lista `frecuencias`, Una vez que se tienen las fecuencias de los resultados se tiene que hacer el histograma. Un histograma es un gráfico que muestra con qué frecuencia se presentan ciertos eventos.

```python
from plotly.graph_objs import Bar, Layout
from plotly import offline

from dado import Dado

miDado = Dado()
resultados = []
for i in range(0,1000,1):
    resultados.append(miDado.tira_dados())

frecuencias = []
for valor in range(1,miDado.num_caras+1):
    frecuencia = resultados.count(valor)
    frecuencias.append(frecuencia)

#Visualización de los resultados
# Genera una lista con valores del 1 al número de caras
valores_x = list(range(1,miDado.num_caras+1))
print(valores_x)
data = [Bar(x=valores_x,y = frecuencias)]

configuracion_eje_x = {'title':'Resultados'}
configuracion_eje_y = {'title':'frecuencia'}

my_layout = Layout(title='Resultados de tirar un dado de 6 caras 1000 veces',xaxis = configuracion_eje_x,
                  yaxis = configuracion_eje_y)

offline.plot({'data':data,'layout':my_layout},filename = 'd6.html')
```

Explicación del código:

1. En el primer `for` se generan los resultados.
2. En el segundo `for` se cuentan cuantas veces ha caido el dado en 1, cuántas veces ha caído el dado en 2 y así sucesivamente.
3. `valores_x = list(range(1,miDado.num_caras+1))`  se genera una lista que tiene los siguientes valores [1,2,3,4,5,6]
4. La clase `Bar()` representa el conjunto de datos para que los presente en un formato de gráfica de barras. Esta clase necesita para el constructor los datos que van sobre el eje `x` y los datos que irán en el eje `y`, que en este caso son `valores_x` y `frecuencias` respectivamente. La clase debe de ir entre corchetes ([]) por que el conjunto de datos puede tener múltiples elementos.
5. Cada elemento se puede configurar de diferentes maneras, en este caso se está utilizando un diccionario donde almacenan el texto que debe de tener los ejes. 
6. La clase `Layout()` gener un objeto donde se almacenan la configuración de la gráfica, en este caso en espcífico el objeto `my_layout` contiene el texto que debe de ir en cada eje y también el título del gráfico
7. Para generar el gráfico, se uitlizará el comando `offline.plot()`. Esta función recibe los datos en forma de diccionario junto con el `Layout`, también acepta el nombre y la extensión en la que se requiere que obtenga el gráfico.

## Trabajando con dos dados
Si se utilizan dos dados, se obtendrá una distribución distinta. Se harán unas modificaciones al código para ahora analizar los eventos generados por utilizar dos dados. 

```python
from plotly.graph_objs import Bar, Layout
from plotly import offline

from dado import Dado

Dado1 = Dado()
Dado2 = Dado()
resultados = []
for i in range(0,1000,1):
    resultados.append(Dado1.tira_dados() + Dado2.tira_dados())

frecuencias = []
for valor in range(2,Dado1.num_caras+Dado2.num_caras+1):
    frecuencia = resultados.count(valor)
    frecuencias.append(frecuencia)

#Visualización de los resultados
# Genera una lista con valores del 1 al número de caras
valores_x = list(range(2,Dado1.num_caras+Dado2.num_caras+1))
print(valores_x)
data = [Bar(x=valores_x,y = frecuencias)]

configuracion_eje_x = {'title':'Resultados','dtick':1}
configuracion_eje_y = {'title':'frecuencia'}

my_layout = Layout(title='Resultados de tirar un dado de 6 caras 1000 veces',xaxis = configuracion_eje_x,
                  yaxis = configuracion_eje_y)

offline.plot({'data':data,'layout':my_layout},filename = 'd6_d6.html')
```
